```python
import pandas as pd
import numpy as np
```

```python
avocado=pd.read_csv("avocado.csv")
avocado.head()
```

```python
avocado.drop(avocado.columns[avocado.columns.str.contains('unnamed', case=False)],
          axis=1, inplace=True)
avocado.head()
```

```python
avocado.index +=1 
avocado.index.name ='S.no'
avocado.head()
```

```python
avocado.dtypes
```

```python
avocado = avocado.rename(columns={'AveragePrice':'Average Price','4046': 'PLU 4046', '4225': 'PLU 4225','4770':'PLU 4770'})
avocado.head()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.dtypes
```

```python
avocado.index = avocado['Date'] 
avocado.head()
```

```python
#plt.figure(figsize=(15,8))
#from matplotlib import rcParams
#rcParams['figure.figsize']=15,8
#avocado.plot(x='Date', y='Average Price')
```

```python
import matplotlib.pyplot as plt
avocado_month=avocado.groupby(avocado['Date'].dt.strftime('%B'))['Average Price'].sum().sort_values()
plt.figure(figsize=(15,8))
avocado_month.plot()
#avocado_month.plot(kind='bar', legend=True, fontsize=12, color ='lightseagreen',rot=45)
plt.title('Avacodo Sales by Month', size=22)
plt.xlabel('Month',size=20)
plt.ylabel('Sales',size=20)
plt.legend(loc=2)
plt.show()
```

```python
avocado_year=avocado.groupby(avocado['Date'].dt.strftime('%Y'))['Average Price'].sum()
plt.figure(figsize=(12,8))
avocado_year.plot(kind='line', legend=True, fontsize=12, color ='rosybrown',rot =45)
plt.title('Avacodo Sales by Year', size=22)
plt.xlabel('Year',size=22)
plt.ylabel('Sales',size=22)
plt.legend(loc=2)
plt.show()
```
